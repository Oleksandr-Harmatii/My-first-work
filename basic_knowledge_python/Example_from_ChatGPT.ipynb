### Змінні та типи даних
Python підтримує кілька основних типів даних, таких як числа, рядки, булеві значення, а також складніші типи, як списки та словники.

```python
# Створення змінних
число = 10         # Ціле число
дробове_число = 3.14  # Число з плаваючою комою
рядок = "Привіт, Python!"  # Рядок
булеве_значення = True     # Булеве значення

print(число, дробове_число, рядок, булеве_значення)
```

### Операції з даними
Python дозволяє виконувати математичні та логічні операції з даними.

```python
# Арифметичні операції
a = 10
b = 3
сума = a + b
різниця = a - b
добуток = a * b
частка = a / b

print("Сума:", сума)
print("Різниця:", різниця)
print("Добуток:", добуток)
print("Частка:", частка)

# Логічні операції
c = True
d = False

print("AND:", c and d)
print("OR:", c or d)
print("NOT c:", not c)
```

### Структури даних
Python має вбудовані структури даних, такі як списки, кортежі, множини та словники, які допомагають організувати дані.

```python
# Список
список = [1, 2, 3, 4]
список.append(5)
print("Список:", список)

# Кортеж
кортеж = (1, 2, 3)
print("Кортеж:", кортеж)

# Множина
множина = {1, 2, 3, 4}
множина.add(5)
print("Множина:", множина)

# Словник
словник = {"ім'я": "Андрій", "вік": 25}
словник["місто"] = "Київ"
print("Словник:", словник)
```

### Умовні конструкції
Python використовує умовні конструкції для управління потоком виконання програми.

```python
вік = 18

if вік >= 18:
    print("Повнолітній")
else:
    print("Неповнолітній")
```

### Цикли
Python має цикл `for` для ітерації через послідовності та цикл `while` для виконання, поки умова істинна.

```python
# Цикл for
for i in range(5):
    print("for:", i)

# Цикл while
лічильник = 0
while лічильник < 5:
    print("while:", лічильник)
    лічильник += 1
```

### Функції
Функції дозволяють створювати блоки коду, які можна використовувати кілька разів.

```python
def привітання(ім_я):
    return f"Привіт, {ім_я}!"

print(привітання("Андрій"))
```
