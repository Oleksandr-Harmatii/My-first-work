```python
import numpy as np
import matplotlib.pyplot as plt

start = 0
stop = 10
step = 1
arange_values = np.arange(start, stop, step)

num = 100 
linspace_values = np.linspace(start, stop, num)

plt.plot(arange_values, label='arange (крок = 1)')
plt.plot(linspace_values, label='linspace (100 елементів)')
plt.legend()

plt.title("Порівняння arange та linspace")
plt.xlabel("Індекс")
plt.ylabel("Значення")

plt.show()
```

###Пояснення
np.arange(start, stop, step) генерує масив, починаючи з start, з кроком step і до значення, меншого за stop. Ви явно задаєте крок.
np.linspace(start, stop, num) генерує масив з num рівномірно розподілених значень між start і stop.
